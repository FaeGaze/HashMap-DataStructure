# HashMap Chaining Resolution

```python
class HashMap:
    def __init__(self):
        self.size = 10
        self.arr = [None] * self.size##[None, None, None, None, None, None, None, None, None, None]

    def _get_hash(self, key):
        hash = 0
        for char in str(key):
                hash += ord(char)
        return hash % self.size ## 1
            
    # Setter for adding Key Value Pairs to the array using chaining methods
    
    def __setitem__(self, key, value):# insert
        #given the key to locate the cell of the array 
        hash_key = self._get_hash(key)
        key_value = [key, value]
        # Boolean Flag
        Correct_key = False
        #Check if that cell is empty or not
        if self.arr[hash_key] is None:##self.arr[5]## not None
            self.arr[hash_key] = list([key_value])
            Correct_key = True
        # cell is not empty, so check if the key exists    
        else:
            #if the key exist, value is updated
            for element in self.arr[hash_key]:
                if len(element) == 2 and element[0] == key:## No
                    element[1] = value
                    Correct_key = True
            #if the key does not exist, the new key value pair is added.        
            if not Correct_key:
                self.arr[hash_key].append(key_value)## If Both have the same key_hash
                Correct_key = True
    
    # Getter To get the value of the key in the specific index
    def __getitem__(self, key):# search
        #given the key to locate the cell of the array 
        key_hash = self._get_hash(key)
        # If the cell is not empty
        if self.arr[key_hash] is not None:
            # Go through the elements in the cell to find the element that matches the key
            for element in self.arr[key_hash]:
                if element[0]==key:
                    return element[1]
            # key is not Found 
            return None
        
    # Deleter to dete the specific key in the specific index             
    def __delitem__(self,key):
        hash_key = self._get_hash(key)
        if self.arr[hash_key] is None:
            return 

        for index, element in enumerate(self.arr[hash_key]):
            if element[0]==key:
                del self.arr[hash_key][index]
                
    # Printing             
    def __str__(self):
        print("Emploee_ID")
        # Print items in the cell that are not None
        print("Hash Table\n", self.arr)
        for item in self.arr:
            if item is not None:
                print("Not Empty Items", str(item))
        

        
if __name__ == '__main__':
    h = HashMap()
    h["Sam Doe"] = 149#1
    h['Bob Lou'] = 197#1
    h['Ming Lou'] = 26#1
    h['Ming Lou'] = 435#1
    
    h["Lisa Smith"] = 27
    h['Ankit Sam'] = 76
    h['Aditya San'] = 852
    h['Alicia Joe'] = 509
    
    h['John Smith'] = 120 #8
    h['Carol Lee 1'] = 130#8
    h["Sandra Dee 5"] = 149#8
    h['Mike Farj'] =  567
    
    print("Ming", h["Ming Lou"])
    h.__str__()
    del h["John Smith"]
    del h["Bob Lou"]
    h.__str__()
```
